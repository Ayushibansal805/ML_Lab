```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("/content/Churn_Modelling.csv")
```

```python
data.info()
```

```python
data.describe()
```

```python
data.drop("RowNumber",axis=1,inplace = True)
```

```python
data.info()
```

```python
data.drop("CustomerId",axis=1,inplace = True)
```

```python
data.info()
```

```python
print(data["Surname"].unique())
print(data["Geography"].unique())
print(data["Gender"].unique())
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data['Surname'] = encoder.fit_transform(data['Surname'])
data['Geography'] = encoder.fit_transform(data['Geography'])
data['Gender'] = encoder.fit_transform(data['Gender'])
```

```python
data.info()
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense,Dropout,BatchNormalization
from tensorflow.keras import regularizers
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.models import Sequential
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
x = data.drop('Exited',axis=1)
y = data['Exited']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.2, random_state = 42)
```

```python
smote =SMOTE(sampling_strategy= 'minority' , random_state = 42)
x_train_sm ,y_train_sm = smote.fit_resample(x_train,y_train)
```

```python
model = Sequential([
    Dense(11,activation= 'relu',input_shape=(x_train.shape[1],),kernel_regularizer = regularizers.L1(0.05)),
    Dense(6,activation = 'relu',kernel_regularizer =regularizers.L2(0.05)),
    Dense(16,activation='relu',kernel_regularizer =regularizers.L1(0.05)),
    Dense(32,activation='relu',kernel_regularizer = regularizers.L2(0.20)),
    BatchNormalization(),
    Dropout(0.3),
    Dense(1,activation ='sigmoid')
])
```

```python
model.compile(optimizer =Adam(),
              loss= 'binary_crossentropy',
              metrics =['accuracy'])
```

```python
history =model.fit(x_train,y_train, epochs =10,batch_size=32, validation_data = (x_test,y_test))
```

```python
plt.plot(history.history['accuracy'], label ='train')
plt.plot(history.history['val_accuracy'],label ='val')

plt.legend()
plt.show()
```
