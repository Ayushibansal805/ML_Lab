```python
from sklearn.decomposition import PCA
model = PCA(n_components=5)
```

```python
y = df['Label']
df.drop('Label', inplace = True, axis = 1)
x = df
```

```python
x_pca = model.fit_transform(x)
```

```python
print(model.explained_variance_ratio_)
```

```python
print(sum(model.explained_variance_ratio_))
```

```python
sns.barplot(y)
```

```python
y.unique()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_pca, y)
```

```python
x_train1 = pd.DataFrame(x_train, columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5'])
```

```python
x_train1.head()
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
```

```python
model1.fit(x_train, y_train)
```

```python
y_pred = model1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score
print(accuracy_score(y_test, y_pred))
print(f1_score(y_test, y_pred))
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=2, copy_x = False, algorithm = "lloyd")
```

```python
kmeans = model.fit(x_train1, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
#silhoutte score
from sklearn.metrics import silhouette_score
silhouette_score(x_train1,kmeans.labels_)
```

```python
from sklearn.cluster import DBSCAN
model = DBSCAN(eps=0.4, min_samples=5, n_jobs=-1)
```
